```python
import nest_asyncio
nest_asyncio.apply()

import os 
from dotenv import load_dotenv

load_dotenv()
```

```python
from pydantic import BaseModel
from pydantic_ai import Agent
from pydantic_ai.common_tools.duckduckgo import duckduckgo_search_tool
from langchain_community.tools import DuckDuckGoSearchRun

class MatchQuery(BaseModel):
    Team1: str
    Team2: str
    Year: int
    league: str

class MatchResult(BaseModel):
    Team1: str
    Team2: str
    Year: int
    league: str
    Scoreline: str
    refereename: str
    stadiumname: str
    found: bool
    winningteam: str


Football_Agent = Agent("groq:qwen/qwen3-32b", instructions="You are a agent, gather the information from the users input like:" \
                        "team1, team2, Year, league. If any of the information is missing then reply more information is need and specify what the user has missed. " \
                        "If you are able to gather all the data, If all details are present, search the internet using DuckDuckGo. " \
                        "Return the result in the required structured format. If you cannot find the match, set found=false. " 
                        ,output_type=[MatchResult],
                        tools=[duckduckgo_search_tool()])


```

```python
result = Football_Agent.run_sync('Who won the match between FC Barcelona and Real Madrid in Laliga in 2026')
print(result.output)
```

```python
result = Football_Agent.run_sync('Who won the match between Arsnel and West Ham United in Premire League in 2026')
print(result.output)
```
