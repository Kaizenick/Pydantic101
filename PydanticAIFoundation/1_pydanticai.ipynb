ASYNCHRONOUS

```python
tranlation_agent = Agent(
    "groq:llama-3.1-8b-instant",
    system_prompt="Translate the sentence into marathi"
)
```

```python
async def get_translation(message: str):
    translation_response = await tranlation_agent.run(message)
    print(translation_response.output)

await get_translation("How are you?")
```

RUN STREAM (ALSO Asynchronous)

```python
joke_agent = Agent(
    "groq:llama-3.1-8b-instant",
    system_prompt="Tell a dad joke"
)
```

```python
async def tell_me_joke(message: str):
    print("Streaming Response.....")
    # We open a stream and listen to the tokens as they arrive
    async with joke_agent.run_stream(message) as result:
        async for word_chunk in result.stream_text(delta=True):
        # We print each word chunk immediately without a newline
            print(word_chunk, end="", flush=True)
    
    print()

await tell_me_joke("GTA San Andreas")
```
