```python
import nest_asyncio
nest_asyncio.apply()

from dotenv import load_dotenv
import os
import subprocess
load_dotenv(verbose=True)
# print(os.getenv("TAVILY_API_KEY")) 
```

```python
from pydantic_ai import Agent
from pydantic_ai.common_tools.tavily import tavily_search_tool

api_key = os.getenv('TAVILY_API_KEY')
```

```python
agent = Agent(
    'groq:llama-3.3-70b-versatile',
    tools=[tavily_search_tool(api_key)],
    system_prompt=(
        "You are a highly capable AI assistant with access to the internet. "
        "Always use your search tool to find up-to-date information before answering."
    )
)
```

```python
result = agent.run_sync('Tell me the top news!')
print(result.output)
```
